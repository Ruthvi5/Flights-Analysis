```python
import pandas as pd

# --- 1) LOAD ---
# adjust paths as needed
airports = pd.read_csv("world_airports.csv")      # OurAirports OR OpenFlights airports.dat saved as csv
routes   = pd.read_csv("routes.dat", header=None)  # OpenFlights routes
routes.columns = ["airline","airline_id","src_code","src_id",
                  "dst_code","dst_id","codeshare","stops","equipment"]

# --- 2) NORMALIZE AIRPORTS SCHEMA ---
def normalize_airports(df):
    cols = {c.lower(): c for c in df.columns}  # case-insensitive helper

    # detect OurAirports vs OpenFlights by key column
    if 'iata_code' in cols or 'latitude_deg' in cols:
        # OurAirports
        return pd.DataFrame({
            'iata': df.get('iata_code'),
            'icao': df.get('ident') if 'ident' in df.columns else df.get('icao_code'),
            'lat':  df.get('latitude_deg'),
            'lon':  df.get('longitude_deg'),
            'name': df.get('name'),
            'country': df.get('iso_country') if 'iso_country' in df.columns else df.get('country')
        })
    else:
        # OpenFlights airports.dat columns (after you applied headers)
        # Expected: IATA, ICAO, Latitude, Longitude, Name, Country
        # If you didn't set headers yet, set them here:
        if list(df.columns) == list(range(len(df.columns))):
            df = df.copy()
            df.columns = ["Airport_ID","Name","City","Country","IATA","ICAO",
                          "Latitude","Longitude","Altitude","Timezone","DST",
                          "Tz_database_time_zone","Type","Source"]
        return pd.DataFrame({
            'iata': df.get('IATA'),
            'icao': df.get('ICAO'),
            'lat':  df.get('Latitude'),
            'lon':  df.get('Longitude'),
            'name': df.get('Name'),
            'country': df.get('Country')
        })

ap = normalize_airports(airports)

# Clean codes to consistent case and strip blanks
for col in ['iata','icao']:
    if col in ap.columns:
        ap[col] = ap[col].astype(str).str.strip().str.upper().replace({'NAN': pd.NA, 'NONE': pd.NA, '': pd.NA})
for col in ['src_code','dst_code']:
    routes[col] = routes[col].astype(str).str.strip().str.upper()

# --- 3) MERGE BY IATA, THEN FALL BACK TO ICAO ---

# Source airport coords: IATA
r = routes.merge(
    ap[['iata','lat','lon']].rename(columns={'iata':'src_code','lat':'lat_src_iata','lon':'lon_src_iata'}),
    on='src_code', how='left'
)
# Source airport coords: ICAO fallback
r = r.merge(
    ap[['icao','lat','lon']].rename(columns={'icao':'src_code','lat':'lat_src_icao','lon':'lon_src_icao'}),
    on='src_code', how='left'
)

# Destination airport coords: IATA
r = r.merge(
    ap[['iata','lat','lon']].rename(columns={'iata':'dst_code','lat':'lat_dst_iata','lon':'lon_dst_iata'}),
    on='dst_code', how='left'
)
# Destination airport coords: ICAO fallback
r = r.merge(
    ap[['icao','lat','lon']].rename(columns={'icao':'dst_code','lat':'lat_dst_icao','lon':'lon_dst_icao'}),
    on='dst_code', how='left'
)

# Coalesce IATA coords with ICAO coords
r['lat_src'] = r['lat_src_iata'].fillna(r['lat_src_icao'])
r['lon_src'] = r['lon_src_iata'].fillna(r['lon_src_icao'])
r['lat_dst'] = r['lat_dst_iata'].fillna(r['lat_dst_icao'])
r['lon_dst'] = r['lon_dst_iata'].fillna(r['lon_dst_icao'])

# Optional: drop helper columns
r = r.drop(columns=[
    'lat_src_iata','lon_src_iata','lat_src_icao','lon_src_icao',
    'lat_dst_iata','lon_dst_iata','lat_dst_icao','lon_dst_icao'
])

# --- 4) CHECK MATCH RATES ---
missing_src = r['lat_src'].isna().sum()
missing_dst = r['lat_dst'].isna().sum()
print(f"Unmatched source airports: {missing_src} | Unmatched destination airports: {missing_dst}")

# Save
r.to_csv("routes_full_with_coords.csv", index=False)
print("Saved routes_full_with_coords.csv")

```

```python
import pandas as pd
import numpy as np

# You already have r (routes merged with OurAirports by IATA/ICAO) and unmatched counts printed.
# Now load OpenFlights airports.dat and normalize columns.

ap_of = pd.read_csv("airports.dat", header=None, dtype=str)
ap_of.columns = ["airport_id","name","city","country","iata","icao",
                 "lat","lon","alt_ft","timezone","dst","tzdb","type","source"]

# Clean types
ap_of["airport_id"] = pd.to_numeric(ap_of["airport_id"], errors="coerce")
for c in ["iata","icao","name","country","city"]:
    ap_of[c] = ap_of[c].astype(str).str.strip().str.upper().replace({"\\N": np.nan, "NAN": np.nan, "": np.nan})
for c in ["lat","lon"]:
    ap_of[c] = pd.to_numeric(ap_of[c], errors="coerce")

# Clean route IDs (routes were named src_id/dst_id earlier)
r["src_id"] = pd.to_numeric(r["src_id"], errors="coerce")
r["dst_id"] = pd.to_numeric(r["dst_id"], errors="coerce")

# --- Fallback for SOURCE by OpenFlights airport_id when lat/lon still missing ---
src_missing_mask = r["lat_src"].isna() | r["lon_src"].isna()
r = r.merge(
    ap_of[["airport_id","lat","lon"]].rename(columns={"airport_id":"src_id","lat":"lat_src_id","lon":"lon_src_id"}),
    on="src_id", how="left"
)
r.loc[src_missing_mask, "lat_src"] = r.loc[src_missing_mask, "lat_src"].fillna(r.loc[src_missing_mask, "lat_src_id"])
r.loc[src_missing_mask, "lon_src"] = r.loc[src_missing_mask, "lon_src"].fillna(r.loc[src_missing_mask, "lon_src_id"])

# --- Fallback for DEST by OpenFlights airport_id when lat/lon still missing ---
dst_missing_mask = r["lat_dst"].isna() | r["lon_dst"].isna()
r = r.merge(
    ap_of[["airport_id","lat","lon"]].rename(columns={"airport_id":"dst_id","lat":"lat_dst_id","lon":"lon_dst_id"}),
    on="dst_id", how="left"
)
r.loc[dst_missing_mask, "lat_dst"] = r.loc[dst_missing_mask, "lat_dst"].fillna(r.loc[dst_missing_mask, "lat_dst_id"])
r.loc[dst_missing_mask, "lon_dst"] = r.loc[dst_missing_mask, "lon_dst"].fillna(r.loc[dst_missing_mask, "lon_dst_id"])

# Drop helper cols
r = r.drop(columns=["lat_src_id","lon_src_id","lat_dst_id","lon_dst_id"])

# Check again
missing_src = r["lat_src"].isna().sum()
missing_dst = r["lat_dst"].isna().sum()
print(f"After OpenFlights-ID fallback -> Unmatched source: {missing_src} | Unmatched destination: {missing_dst}")

r.to_csv("routes_full_with_coords.csv", index=False)
print("Re-saved routes_full_with_coords.csv")
```

```python
# Load OurAirports with all relevant keys (if not already loaded)
ap_oa = pd.read_csv("world_airports.csv")  # OurAirports
# Normalize
def clean_code(s):
    return s.astype(str).str.strip().str.upper().replace({'NAN': pd.NA, 'NONE': pd.NA, '': pd.NA})

for c in ['iata_code','ident','local_code','iso_country','name','municipality']:
    if c in ap_oa.columns:
        ap_oa[c] = clean_code(ap_oa[c])

ap_oa['lat'] = pd.to_numeric(ap_oa['latitude_deg'], errors='coerce')
ap_oa['lon'] = pd.to_numeric(ap_oa['longitude_deg'], errors='coerce')

# FAA/local-code fallback (common in US routes)
# SOURCE by local_code
r = r.merge(
    ap_oa[['local_code','lat','lon']].rename(columns={'local_code':'src_code','lat':'lat_src_loc','lon':'lon_src_loc'}),
    on='src_code', how='left'
)
src_missing_mask = r['lat_src'].isna() | r['lon_src'].isna()
r.loc[src_missing_mask, 'lat_src'] = r.loc[src_missing_mask, 'lat_src'].fillna(r.loc[src_missing_mask, 'lat_src_loc'])
r.loc[src_missing_mask, 'lon_src'] = r.loc[src_missing_mask, 'lon_src'].fillna(r.loc[src_missing_mask, 'lon_src_loc'])

# DEST by local_code
r = r.merge(
    ap_oa[['local_code','lat','lon']].rename(columns={'local_code':'dst_code','lat':'lat_dst_loc','lon':'lon_dst_loc'}),
    on='dst_code', how='left'
)
dst_missing_mask = r['lat_dst'].isna() | r['lon_dst'].isna()
r.loc[dst_missing_mask, 'lat_dst'] = r.loc[dst_missing_mask, 'lat_dst'].fillna(r.loc[dst_missing_mask, 'lat_dst_loc'])
r.loc[dst_missing_mask, 'lon_dst'] = r.loc[dst_missing_mask, 'lon_dst'].fillna(r.loc[dst_missing_mask, 'lon_dst_loc'])

# Clean helper columns
r = r.drop(columns=['lat_src_loc','lon_src_loc','lat_dst_loc','lon_dst_loc'])

# Recount after local_code fallback
missing_src = r['lat_src'].isna().sum()
missing_dst = r['lat_dst'].isna().sum()
print(f"After local_code fallback -> Unmatched source: {missing_src} | Unmatched destination: {missing_dst}")
```
